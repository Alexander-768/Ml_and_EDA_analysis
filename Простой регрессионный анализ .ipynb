Сначала импортиуем все


```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import scipy.stats as stats
```

Создаем датафрейм

```python
def create_salary_dataframe():
    return pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
salary = create_salary_dataframe()
```

Чуть обработаем его


```python
salary.columns = ['Experience_Years', 'Salary']
print(salary)
```

Проверяем условиие линейной и монотонной взаимосвязи X и Y


```python
plt.plot(salary['Experience_Years'], salary['Salary'], 'o')
plt.show()
```

На графике заметим что выполнилось условие линейной и монотонной зависимости зависимой переменной от независимой


Построим модель регрессионного анализа


```python
model = smf.ols('Salary ~ Experience_Years', data=salary).fit()
```

Проводим проверку нормальности распределения остатков через QQ plot

```python
residuals =model.resid
stats.probplot(residuals, dist="norm", plot=plt)
plt.show()
```


**Получим мат модель**

```python
mat_model=(model.params)
r_squared = model.rsquared
print(f"Математическая модель: y = {round(mat_model['Intercept'], 3)} + {round(mat_model['Experience_Years'], 3)}x с коэфицентом детерминации {round(r_squared, 3)}")
```
